This *Jupyter Notebook* contains the code (UNIX-based shell commands and python scripts) needed to train (locally or on a hosted environment on the cloud) a neural net using Tensorflow's official *Object detection API*. Please execute all these instructions in order.

In the example used in this *Jupyter Notebook*, we train the neural net using the configurations of a **faster_rcnn_inception_v2_coco** object detection model and the pre-trained model found in this [link](http://download.tensorflow.org/models/object_detection/ssd_mobilenet_v1_fpn_shared_box_predictor_640x640_coco14_sync_2018_07_03.tar.gz).

SSDs configurations and pretrained models found in the official *Object detection API* (such as the **ssd_mobilenet_v1_coco** and the **ssd_mobilenet_v1_fpn**) were originally used, but they caused incomprehensible exceptions and errors (such as *incompatible shapes error*). Hence, it is not recommended to use SSD-based models and configurations unless one understands how they work.
